```python
# خواندن داده‌ها از فایل
data = []
with open('/content/drive/MyDrive/HardwareAcceleraation/Dataset/Customers1.csv', 'r') as file:
    for line in file:
        row = line.strip().split(',')
        data.append(row)

# حذف سرتیتر
header = data[0]
data = data[1:]

# تبدیل داده‌ها به اعداد
for row in data:
    row[0] = int(row[0])
    row[2] = int(row[2])
    row[3] = float(row[3])
    row[4] = float(row[4])

# انتخاب ویژگی‌های ورودی برای KMeans
feature1_index = header.index('Annual Income')
feature2_index = header.index('Spending Score')
X = [[row[feature1_index], row[feature2_index]] for row in data]

# پیاده‌سازی الگوریتم KMeans
class KMeans:
    def __init__(self, n_clusters):
        self.n_clusters = n_clusters

    def fit(self, X):
        # ابتدا مرکزهای اولیه را به صورت تصادفی انتخاب می‌کنیم
        centroids = [X[i] for i in range(self.n_clusters)]

        # الگوریتم KMeans
        for _ in range(100):  # حداکثر تعداد تکرارها
            clusters = [[] for _ in range(self.n_clusters)]
            for point in X:
                distances = [((point[0] - centroid[0]) ** 2 + (point[1] - centroid[1]) ** 2) ** 0.5 for centroid in centroids]
                cluster_index = distances.index(min(distances))
                clusters[cluster_index].append(point)

            new_centroids = []
            for cluster in clusters:
                if cluster:
                    new_centroid = [sum(point[0] for point in cluster) / len(cluster), sum(point[1] for point in cluster) / len(cluster)]
                    new_centroids.append(new_centroid)
                else:
                    # در صورتی که خوشه‌ای خالی باشد، مرکز قبلی را نگه می‌داریم
                    new_centroids.append(centroids[new_centroids.index(cluster)])
            if new_centroids == centroids:
                break
            centroids = new_centroids
        self.labels_ = []
        for point in X:
          distances = [((point[0] - centroid[0]) ** 2 + (point[1] - centroid[1]) ** 2) ** 0.5 for centroid in centroids]
          cluster_index = distances.index(min(distances))
          self.labels_.append(cluster_index)

        self.cluster_centers_ = centroids

# اعمال الگوریتم KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
clusters = kmeans.labels_
centers = kmeans.cluster_centers_

# گزارش خروجی
output = [['CustomerID', 'Cluster', 'Cluster Center']]
for i, row in enumerate(data):
    output.append([row[0], clusters[i], centers[clusters[i]]])

# نوشتن خروجی به فایل csv در مسیر مورد نظر
import csv
output_path = "/content/drive/MyDrive/HardwareAcceleraation/Dataset/output.csv"
with open(output_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(output)
```
